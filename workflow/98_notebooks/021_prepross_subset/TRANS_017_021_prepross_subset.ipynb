```python
from pathlib import Path
import scanpy as sc
```

```python
subset_dir = Path("/srv/ext_home/asbarros/projects/BSSantos/TRANS_017/data/021_preprocess_subset")
subset_dir.mkdir(parents=True, exist_ok=True)
```

### Sample 1 Subsetting

```python
SAMPLE_NAME = "Sample_1"

cellranger_sample1 = Path("/srv/ext_home/asbarros/ifs/ADA/SHARED/projects/BSSantos/TRANS_017/data/01_cellranger_output/Sample_1/outs/")

adata_s1 = sc.read_10x_mtx(
    cellranger_sample1 / "filtered_feature_bc_matrix",
    cache=True,gex_only=False
)

adata_s1_subset = sc.pp.subsample(adata_s1, n_obs=500, copy=True, random_state=42)

output_file = subset_dir/ f"{SAMPLE_NAME}_subset.h5ad"
adata_s1_subset.write(output_file)
```
